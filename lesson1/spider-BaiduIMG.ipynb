```python
import requests
import os
import imghdr
```

```python
def getMorePages(keyword, pages):
    params=[]
    for i in range(30, 30*pages+30, 30): #步长为30
        params.append({
            'tn': 'resultjson_com',
            'ipn': 'rj',
            'ct': '201326592',
            'is':'' ,
            'fp': 'result',
            'queryWord': keyword,
            'cl': 2,
            'lm': -1,
            'ie': 'utf-8',
            'oe': 'utf-8',
            'adpicid': '',
            'st': -1,
            'z': '',
            'ic': '',
            'hd': '',
            'latest': '',
            'copyright': '',
            'word': keyword,
            's': '',
            'se': '',
            'tab': '',
            'width': '',
            'height': '',
            'face': 0,
            'istype': 2,
            'qc': '',
            'nc': 1,
            'fr': '',
            'expermode': '',
            'force': '',
            'pn': i,
            'rn': 30,
            'gsm': '1e',})
    url = 'https://image.baidu.com/search/index'
    urls = []
    for i in params:
        res = requests.get(url, params=i)
        for item in res.json()['data'][:-1]:
            urls.append(item['thumbURL'])
    return urls
```

```python
def writeImg(Path, urls):
    if not os.path.exists(Path): 
        os.mkdir(Path)
    i = 1;
    for u in urls:
        img = requests.get(u)
        with open(Path+'/'+str(i)+'.jpg', 'wb') as f:
            f.write(img.content)
            f.close()
        if imghdr.what(Path+'/'+str(i)+'.jpg')=='jpeg': #判断是否损坏
            print(str(i)+'.jpg is downloaded')
            i+=1
```

```python
def crawl(Path, key, page=20):
    urls = getMorePages(key,page)
    writeImg(Path, urls)
```

```python
crawl('data/schoolbus','编程')
```
