```python
from fastai.vision import *
from fastai.metrics import error_rate
from PIL import Image
```

# Train a model to classify different types of cars with resNet34

## 1 Preparation
I've collected approximately 5 * 600 jpg pictures with defferent shapes, and made a ImageNet style dataset from them.  

```python
path = Path('data')
```

```python
(path/'train').ls()
```

```python
data = ImageDataBunch.from_folder(path, ds_tfms=get_transforms(), size=224)
```

```python
data.normalize(imagenet_stats)
```

```python
data.show_batch(rows=3, figsize=(7,6))
```

seems not bad!!!!  fast.ai's normalization is amazing!

```python
print(data.classes)
```

## 2 Training: resnet34

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
learn.model
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

## 3 check

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_top_losses(9, figsize=(10,10))
```

What a shame! Almost all the mistakes it makes are on images which are wrong originally. I'm not a good data collector but resnet34 is an excellent learner.

```python
interp.plot_confusion_matrix(figsize=(4,4),dpi=100)
```

### Trick of `ImageDataBunch.from_folder()`
Set `valid_pct`= `proportion` of valid set in your whole dataset, and it will automatically split into two part. Don't need to split it by hand.
